```python
#merging the constructors dataframe with race results

team = constructors.merge(results,on='constructorId',how = 'left')
```

```python
team
```

```python
#extracting the columns needed and grouping it by constructor name, extracting the total races entered

best = team[['name','points','raceId']]
best = best.groupby('name')['raceId'].nunique().sort_values(ascending=False).reset_index(name = 'races')
best = best[best['races'] >= 100]
best.head() 
```

```python
#building a formula to calculate points per race 

func = lambda x: x.points.sum()/x.raceId.nunique()
data = team[team['name'].isin(best.name)].groupby('name').apply(func).sort_values(ascending=False).reset_index(name = 'points_per_race')
data.head(10)
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
map_center = [circuits['lat'].mean(), circuits['lng'].mean()]
world_map = folium.Map(location=map_center, zoom_start=2)


marker_cluster = MarkerCluster().add_to(world_map)


for index, row in circuits.iterrows():
    folium.Marker(
        location=[row['lat'], row['lng']],
        popup=f"{row['name']}, {row['country']}",
        icon=folium.Icon(color='blue')
    ).add_to(marker_cluster)


world_map
```

```python
# Plot the top 10 races
plt.figure(figsize=(10, 6))
top_10.plot(kind='bar', color='skyblue')
plt.xlabel('Circuit Names')
plt.ylabel('Count')
plt.title('Top 10 Races by Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
num_races_by_year = races['year'].value_counts().reset_index()
```

```python
num_races_by_year
```

```python
# Rename columns 'index' and 'year' for index you'll change it to year and year to races_per_season
num_races_by_year.rename(columns={'index': 'year', 'year': 'races_per_season'}, inplace=True)
```

```python
num_races_by_year
```

```python
num_races_by_year_sort = num_races_by_year.sort_values(by='year', ascending=True).reset_index(drop=True)
```

```python
num_races_by_year_sort
```

```python
# Create a bar plot for the number of races per season
plt.figure(figsize=(10, 6))
plt.bar(num_races_by_year_sort['year'], num_races_by_year_sort['races_per_season'], color='skyblue')
plt.xlabel('Year')
plt.ylabel('Count of Races')
plt.title('Number of Races per season')
plt.grid(axis='y')
plt.show()
```
